## 4. FEATURE ENGINEERING

```python
print("\n" + "="*5)
print("FEATURE ENGINEERING")
print("="*5)

def engineer_features(df):
    """Apply feature engineering to the dataset"""
    
    # Create a copy to avoid modifying original
    df = df.copy()
    
    # 1. Extract Title from Name
    df['Title'] = df['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
    
    # Group rare titles
    title_mapping = {
        'Mr': 'Mr', 'Miss': 'Miss', 'Mrs': 'Mrs', 'Master': 'Master',
        'Dr': 'Rare', 'Rev': 'Rare', 'Col': 'Rare', 'Major': 'Rare',
        'Mlle': 'Miss', 'Ms': 'Miss', 'Mme': 'Mrs',
        'Don': 'Rare', 'Dona': 'Rare', 'Lady': 'Rare', 'Countess': 'Rare',
        'Jonkheer': 'Rare', 'Sir': 'Rare', 'Capt': 'Rare'
    }
    df['Title'] = df['Title'].map(title_mapping)
    df['Title'] = df['Title'].fillna('Rare')
    
    # 2. Family Size and Is Alone
    df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
    df['IsAlone'] = (df['FamilySize'] == 1).astype(int)
    
    # 3. Fare per person
    df['FarePerPerson'] = df['Fare'] / df['FamilySize']
    
    # 4. Age categories
    df['AgeCat'] = pd.cut(df['Age'], bins=[0, 12, 18, 35, 60, 100], 
                           labels=['Child', 'Teen', 'Adult', 'MiddleAge', 'Senior'])
    
    # 5. Fare categories
    df['FareCat'] = pd.qcut(df['Fare'].fillna(df['Fare'].median()), q=4, 
                             labels=['Low', 'Medium', 'High', 'VeryHigh'])
    
    # 6. Cabin deck (first letter of cabin)
    df['Deck'] = df['Cabin'].str[0]
    df['HasCabin'] = df['Cabin'].notna().astype(int)
    
    return df
print("New features created:")
print("- Title (extracted from Name)")
print("- FamilySize (SibSp + Parch + 1)")
print("- IsAlone (binary)")
print("- FarePerPerson")
print("- AgeCat (age categories)")
print("- FareCat (fare categories)")
print("- Deck (from Cabin)")
print("- HasCabin (binary)")

# Apply feature engineering
train_featured = engineer_features(train_df)
test_featured = engineer_features(test_df)
```

## 5. DATA PREPROCESSING

```python
print("\n" + "="*5)
print("DATA PREPROCESSING")
print("="*5)

def preprocess_data(train, test):
    """Preprocess training and test data"""
    
    # Combine for consistent preprocessing
    combined = pd.concat([train, test], sort=False)
    
    # Fill missing Age values using Title groups
    age_by_title = combined.groupby('Title')['Age'].median()
    for title in age_by_title.index:
        combined.loc[(combined['Title'] == title) & (combined['Age'].isnull()), 'Age'] = age_by_title[title]
    
    # Fill missing Embarked (only in train)
    combined['Embarked'].fillna(combined['Embarked'].mode()[0], inplace=True)
    
    # Fill missing Fare (only one in test)
    combined['Fare'].fillna(combined['Fare'].median(), inplace=True)
    
    # Recreate age and fare categories after filling
    combined['AgeCat'] = pd.cut(combined['Age'], bins=[0, 12, 18, 35, 60, 100], 
                                 labels=['Child', 'Teen', 'Adult', 'MiddleAge', 'Senior'])
    combined['FareCat'] = pd.qcut(combined['Fare'], q=4, 
                                   labels=['Low', 'Medium', 'High', 'VeryHigh'])
    combined['FarePerPerson'] = combined['Fare'] / combined['FamilySize']
    
    # Encode categorical variables
    categorical_features = ['Sex', 'Embarked', 'Title', 'AgeCat', 'FareCat', 'Deck']
    
    for feature in categorical_features:
        le = LabelEncoder()
        combined[feature] = le.fit_transform(combined[feature].astype(str))
    
    # Select features for modeling
    features = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 
                'Embarked', 'Title', 'FamilySize', 'IsAlone', 
                'FarePerPerson', 'AgeCat', 'FareCat', 'HasCabin']
    
    # Split back
    train_processed = combined[:len(train)][features]
    test_processed = combined[len(train):][features]
    
    return train_processed, test_processed

# Preprocess data
X_train, X_test_final = preprocess_data(train_featured, test_featured)
y_train = train_df['Survived']

print(f"Training features shape: {X_train.shape}")
print(f"Test features shape: {X_test_final.shape}")
print(f"\nFeatures used: {list(X_train.columns)}")
```

## 6. MODEL TRAINING

```python
print("\n" + "="*5)
print("MODEL TRAINING")
print("="*5)

# Split for validation
X_tr, X_val, y_tr, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42, stratify=y_train)

# Scale features
scaler = StandardScaler()
X_tr_scaled = scaler.fit_transform(X_tr)
X_val_scaled = scaler.transform(X_val)
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test_final)

# Initialize models
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'Gradient Boosting': GradientBoostingClassifier(n_estimators=100, random_state=42),
    'SVM': SVC(probability=True, random_state=42),
    'KNN': KNeighborsClassifier(n_neighbors=5)
}

# Train and evaluate models
results = {}
best_score = 0
best_model = None

print("\nModel Performance (Cross-Validation):")
print("-" * 40)

for name, model in models.items():
    # Use cross-validation
    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    scores = cross_val_score(model, X_train_scaled, y_train, cv=cv, scoring='accuracy')
    
    results[name] = {
        'mean': scores.mean(),
        'std': scores.std(),
        'scores': scores
    }
    
    print(f"{name:20} | Accuracy: {scores.mean():.4f} (+/- {scores.std():.4f})")
    
    if scores.mean() > best_score:
        best_score = scores.mean()
        best_model = model

print(f"\nBest Model: {best_model.__class__.__name__} with accuracy: {best_score:.4f}")
```

## 7. ENSEMBLE MODEL

```python
print("\n" + "="*5)
print("CREATING ENSEMBLE MODEL")
print("="*5)

# Create ensemble of best performers
ensemble = VotingClassifier(
    estimators=[
        ('lr', LogisticRegression(max_iter=1000, random_state=42)),
        ('rf', RandomForestClassifier(n_estimators=200, max_depth=10, random_state=42)),
        ('gb', GradientBoostingClassifier(n_estimators=100, max_depth=4, random_state=42))
    ],
    voting='soft'
)

# Train ensemble
ensemble.fit(X_train_scaled, y_train)

# Evaluate ensemble
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
ensemble_scores = cross_val_score(ensemble, X_train_scaled, y_train, cv=cv, scoring='accuracy')
print(f"Ensemble Model Accuracy: {ensemble_scores.mean():.4f} (+/- {ensemble_scores.std():.4f})")
```
